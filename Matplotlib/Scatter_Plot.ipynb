```python
import matplotlib.pyplot as plt
x=[5,10,17,20,25,2,35]
y=[2,4,6,8,10,12,14]
colors=['r','y','g','b','r','g','r']
sizes=[100,250,150,300,75,100,600]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,edgecolor="black",linewidth=2)
plt.colorbar()
plt.title("Sales",fontsize=15)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Growth",fontsize=15)
plt.show()
```
