##### Bar Plot

```python
import matplotlib.pyplot as plt
x=["C","C++","Python","Java"]
y=[80,40,85,90]
plt.xlabel("languages",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.bar(x,y,width=0.2,color="b",edgecolor="r",alpha=0.5,linestyle=":",linewidth=2,label="popularity")
plt.legend()
plt.title("Language Graph")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["C","C++","Python","Java"]
y=[80,40,85,90]
z=[20,30,10,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=15)
plt.ylabel("No.",fontsize=15)

plt.bar(x,y,width,color="b",alpha=0.5,label="popularity")
plt.bar(p1,z,width,color="r",alpha=0.5,label="popularity1")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["C","C++","Python","Java"]
y=[80,40,85,90]
z=[20,30,10,50]

plt.xlabel("languages",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Language Graph")
plt.barh(x,y,width,color="b",alpha=0.5,label="popularity")
plt.barh(p1,z,width,color="r",alpha=0.5,label="popularity1")

plt.legend()
plt.show()
```
